```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import os

data = pd.read_csv('games.csv')
print(data)
```

```python
#결측치(Missing data)
print('전체 데이터 건수:', len(data))
```

```python
print('칼럼별 결측치 개수')
len(data) - data.count()
```

```python
# item-based 분석 데이터 칼럼 중 전체 데이터 건 수 대비 상당 수가 
# 결측치로 나와있는 칼럼인 Reviews, Website 등은 삭제한다.
# 또한 분석에 의미 없다고 판단되는 데이터도 삭제한다.
data1 = data.drop(['About the game','Reviews','Header image','Website','Support url','Support email','Metacritic url','Notes','Screenshots'], axis = 'columns')
data1.head()
```

```python
# 6개의 열이 삭제되고 총 33개의 칼럼이 남아있다.
# 이제 Primary key로 삼을 AppID에 중복된 데이터가 있는지 확인한다.
data1.duplicated()
```
