```python
try:
    import pandas as pd

    # 파일을 읽고 데이터프레임을 만듭니다.
    data = pd.read_csv('../Dataset/data1.csv')

    # 필요한 컬럼만 선택
    selected_columns = ['AppID', 'Supported languages', 'Full audio languages', 'Price']
    subset_data = data[selected_columns]

    # 새로운 CSV 파일로 저장
    #subset_data.to_csv('subset_data1.csv', index=False)

except FileNotFoundError as e:
    print("Error: Cannot find the File(CSV), Check the Path again")
    exit(1)
```

original dataset: games.csv(경호님)

```python
try:
    import pandas as pd

    # 파일을 읽고 데이터프레임을 만듭니다.
    subdata = pd.read_csv('../Dataset/subset_data1.csv')
    print(subdata)
except FileNotFoundError as e:
    print("Error: Cannot find the File(CSV), Check the Path again")
    exit(1)
```

```python
subdata = subdata.merge(subset_data, on='AppID', how='inner')
print(subdata)
```

ENCODING & Scaling

1. One Hot Encoding

Features: Supported languages, Full audio languages 

```python
# One hot Encdoing for Supported languages Full audio languages 
supported_languages = subdata['Supported languages'].str.strip('[]').str.split(', ').str.join('|').str.get_dummies()
full_audio_languages = subdata['Full audio languages'].str.strip('[]').str.split(', ').str.join('|').str.get_dummies()

supported_languages = supported_languages.add_prefix('Supported_languages_')
full_audio_languages = full_audio_languages.add_prefix('Full_audio_languages_')

# merge
encoded_data = pd.concat([subdata, supported_languages, full_audio_languages], axis=1)

# delete original
encoded_data = encoded_data.drop(['Supported languages', 'Full audio languages'], axis=1)
print(encoded_data)
```

2. Scaling: Min-Max Scaling

Features: Price, required_age

```python
from sklearn.preprocessing import MinMaxScaler

# Min-Max 스케일러 객체 생성
scaler = MinMaxScaler()

# Price 및 required_age 컬럼 스케일링
encoded_data[['Price', 'required_age']] = scaler.fit_transform(encoded_data[['Price', 'required_age']])
print(encoded_data)
```

Final preprocessed Dataset for Clustering

```python
encoded_data.to_csv('cluster_data.csv', index=False)
print(encoded_data)
```
